```python
from google.colab import auth
from google.cloud import bigquery
import pandas as pd
```

```python
auth.authenticate_user()
project_id = 'poised-honor-384014'  # Substitua pelo ID do seu projeto do Google Cloud
client = bigquery.Client(project=project_id)
```

```python
query_exchanges = '''
SELECT *
FROM `poised-honor-384014.home_exchange.exchanges` limit 1000000
'''
```

```python
exchanges = client.query(query_exchanges).to_dataframe()
```

```python
query = '''
SELECT *
FROM `poised-honor-384014.home_exchange.subscriptions`
'''
```

```python
subscriptions = client.query(query).to_dataframe()
subscriptions #Tabela subscriptions bruta
```

```python
subscriptions_sem_duplicatas = subscriptions.drop_duplicates()
subscriptions_sem_duplicatas #Tiramos as linhas duplicadas
```

```python
subscriptions_sem_duplicatas['renew_count'] = subscriptions_sem_duplicatas.groupby('user_id')['renew'].transform('count')
subscriptions_sem_duplicatas.head() #Contamos o número de renew por user_id
```

```python
not_renew_subscriptions = subscriptions_sem_duplicatas[subscriptions_sem_duplicatas['renew'] == 0]
not_renew_subscriptions #Filtramos as subscriptions que não foram renovadas
```

```python
renew_subscriptions = subscriptions_sem_duplicatas[subscriptions_sem_duplicatas['renew'] == 1]
renew_subscriptions #Filtramos as subscriptions que foram renovadas
```

```python
renew_subscriptions_df = pd.merge(renew_subscriptions, exchanges, left_on='user_id', right_on='creator_id', how='inner')
renew_subscriptions_df[['subscription_date','user_id','exchanges_count']].groupby('user_id')
renew_subscriptions_df #Unimos as sbscriptions que renovaram com a tabela de exchanges feitas.
```

```python
renew_subscriptions_df['total_exchanges'] = renew_subscriptions_df.groupby('user_id')['exchange_id'].transform('count')
renew_subscriptions_df #Calculamos o número de exchanges feitas pelos usuários que renovaram
```

```python
not_renewed_merged = pd.merge(not_renew_subscriptions, exchanges, left_on='user_id', right_on='creator_id', how='inner')
not_renewed_merged['total_exchanges'] = not_renewed_merged.groupby('user_id')['exchange_id'].transform('count')
not_renewed_merged #Unimos as subscriptions que não renovaram com a tabela de exchanges feitas.  #Calculamos o número de exchanges feitas pelos usuários que não renovaram
```

```python
exchanges['exchanges_count'] = exchanges.groupby('creator_id')['exchange_id'].transform('count')
exchanges.to_csv('exchanges_python.csv', index=False)
```

```python
subscriptions_exchanges = pd.merge(subscriptions_sem_duplicatas, exchanges, left_on='user_id', right_on='creator_id', how='inner')
subscriptions_exchanges.to_excel('subscriptions_exchanges.xlsx', index=False) #Unimos as tabelas subscriptions com exchange.
```
